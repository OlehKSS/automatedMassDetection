## Reading files

This example shows how to read one image and create MammogramImage file

This example shows how to read whole dataset

```python
from immas.io import read_dataset
import cv2

data_set = read_dataset(image_folder="C:\dataset\images",
            mask_folder="C:\dataset\masks",
            results_folder="C:\dataset\groundtruth",
            pmuscle_mask_folder="C:\dataset\pectoral_muscle_masks",
            train_set_fraction=1)

print("Number of images for training is {0}, number of images for testing is {1}".format(
    len(data_set["train"]), len(data_set["test"])))
```

```python
# Segment and shows one mass of the dataset

from skimage import measure

img = data_set['train'][0]
img.read_data()
#plt.imshow(img.image_ground_truth, cmap="gray")
props = measure.regionprops(img.image_ground_truth)
region = img.uncropped_image[props[0].bbox[0]:props[0].bbox[2],props[0].bbox[1]:props[0].bbox[3]]
plt.imshow(region, cmap="gray")
```

```python
# Segment and save each mass of the dataset

from skimage import measure
import cv2
import os

savepath = "C:\dataset\extracted_masses"

for img in data_set['train']:
    img.read_data()
    labeled = measure.label(img.image_ground_truth, neighbors=8)
    props = measure.regionprops(labeled)
    for r in range(0, len(props)):
        region = img.uncropped_image[props[r].bbox[0]:props[r].bbox[2],props[r].bbox[1]:props[r].bbox[3]]
        name = img.file_name + str(r)
        cv2.imwrite(os.path.join(savepath, name) + '.tif', region)
```

## Showing Images using OpenCV or Matplotlib

```python
from scipy import ndimage, misc
import numpy as np
import os
import cv2
from matplotlib import pyplot
import imageio

def main():
    outPath = "C:\\automatedMassDetection\\dataset\\e"
    path = "C:\\automatedMassDetection\\dataset\\extracted_masses_Copia"

    # iterate through the names of contents of the folder
    for image_path in os.listdir(path):

        # create the full input path and read the file
        input_path = os.path.join(path, image_path)
        image_to_rotate = matplotlib.pyplot.imread(input_path)

        # rotate the image
        rotated = ndimage.rotate(image_to_rotate, 180)

        # create full output path, 'example.jpg' 
        # becomes 'rotate_example.jpg', save the file to disk
        fullpath = os.path.join(outPath, 'rotated_'+image_path)
        imageio.imwrite(fullpath, rotated)

if __name__ == '__main__':
    main()
```
